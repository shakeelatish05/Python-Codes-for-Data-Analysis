```python
import pandas as pd
df=pd.read_csv('Strokes.csv')
df.head()
```

```python
df.describe()
```

```python
#Perform Chi Square test two assess association between worktype and Stroke
from scipy.stats import chi2_contingency
```

```python
 #Create a contingency table of the 'work_type' and 'stroke' columns
contingency_table = pd.crosstab(df['work_type'], df['stroke'])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
```

```python
# Odd ratio of stroke

import numpy as np

# Calculate the odds ratio for 'Private' work type having a stroke
# contingency_table is already defined from the preceding code
odds_private_stroke = (contingency_table.loc['Private', 1] / contingency_table.loc['Private', 0])
odds_other_stroke = (contingency_table.sum(axis=0)[1] - contingency_table.loc['Private', 1]) / (contingency_table.sum(axis=0)[0] - contingency_table.loc['Private', 0])

odds_ratio = odds_private_stroke / odds_other_stroke
print(f"Odds Ratio for Private work type having a stroke: {odds_ratio}")
print(f"The odds of having a stroke in private group is  {100*(odds_ratio-1)} % higher in the private group as compared to others")

```

```python
# You can generalize this for other work types if needed
odds_self_employed_stroke = (contingency_table.loc['Self-employed', 1] / contingency_table.loc['Self-employed', 0])
odds_other_stroke_excluding_self_employed = (contingency_table.sum(axis=0)[1] - contingency_table.loc['Self-employed', 1]) / (contingency_table.sum(axis=0)[0] - contingency_table.loc['Self-employed', 0])
odds_ratio_self_employed = odds_self_employed_stroke / odds_other_stroke_excluding_self_employed
print(f"Odds Ratio for Self-employed work type having a stroke: {odds_ratio_self_employed}")
```

```python
# Interpret the results
alpha = 0.05
if p < alpha:
    print("Result: Reject the null hypothesis. There is a significant association between work type and stroke.")
else:
    print("Result: Fail to reject the null hypothesis. There is no significant association between work type and stroke.")
```

```python
#Use Fisher's Exact Test between residence and strok

import pandas as pd
from scipy.stats import fisher_exact

# Create a contingency table of the 'residence_type' and 'stroke' columns
contingency_table_residence_stroke = pd.crosstab(df['Residence_type'], df['stroke'])

# Perform Fisher's Exact Test
oddsratio, p_value = fisher_exact(contingency_table_residence_stroke)

# Print the results
print(f"Odds Ratio (Fisher's Exact Test): {oddsratio}")
print(f"P-value (Fisher's Exact Test): {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. There is a significant association between residence type and stroke.")
else:
    print("Result: Fail to reject the null hypothesis. There is no significant association between residence type and stroke.")
```

```python
from scipy.stats import binomtest

x = 60  # Number of successes (heads)
n = 100 # Total trials (flips)
p = 0.5 # Hypothesized probability of success (fair coin)

result = binomtest(x, n, p, alternative='two-sided')

print(f"P-value: {result.pvalue}")

if result.pvalue < 0.05:
    print("Reject the null hypothesis: The coin is likely not fair.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to say the coin is unfair.")
```
