```python
import pandas as pd
df=pd.read_csv('Strokes.csv')
df.head()
```

```python
# prompt: Provide n, mean and standard deviation of glucose level across Work-type groups

df[['work_type','avg_glucose_level']].groupby('work_type').agg(['count','mean', 'std'])
```

```python
# prompt: One way Anova to compare glucose level across Worktype groups
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Perform one-way ANOVA
model = ols('avg_glucose_level ~ C(work_type)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

anova_table
```

```python
# prompt: Post hoc test Tukey HSD test
from statsmodels.stats.multicomp import pairwise_tukeyhsd
# perform Tukey's post hoc test
tukey_result = pairwise_tukeyhsd(endog=df['avg_glucose_level'], groups=df['work_type'], alpha=0.05)

# print the results
print(tukey_result)
```
