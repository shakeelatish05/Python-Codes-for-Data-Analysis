```python
import pandas as pd
df=pd.read_csv('Strokes.csv')
df.head()
```

```python
# prompt: Pie chart

import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.pie(df['stroke'].value_counts(),labels=['No','Yes'],autopct='%1.2f%%')
plt.title('Stroke Distribution')
plt.show()
```

```python
# prompt: bar plot for work_type

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
df['work_type'].value_counts().plot(kind='bar')
plt.title('Distribution of Work Types')
plt.xlabel('Work Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: Multiple bar char for work_type and gender

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='work_type', hue='gender')
plt.title('Distribution of Work Types by Gender')
plt.xlabel('Work Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# prompt: Stack bar chart for work type and gender

import matplotlib.pyplot as plt
# To create a stacked bar chart, we can use pandas groupby and unstack
work_gender_counts = df.groupby(['work_type', 'gender']).size().unstack()

plt.figure(figsize=(10, 6))
work_gender_counts.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Stacked Bar Chart of Work Type by Gender')
plt.xlabel('Work Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# prompt: Scatter plot matrix for age bmi glocose level etc

import matplotlib.pyplot as plt
# Select the relevant numerical columns
numerical_cols = ['age', 'bmi', 'avg_glucose_level']

# Create a scatter plot matrix
sns.pairplot(df)
plt.suptitle('Scatter Plot Matrix of Numerical Features', y=1.02) # Add a title
plt.show()
```

```python
# prompt: Heat map

import matplotlib.pyplot as plt
# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
