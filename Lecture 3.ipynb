```python
import pandas as pd
df=pd.read_csv('Strokes.csv')
df.head()
```

```python
#Missing values in different variables

df.isnull().sum()
```

```python
# Rreplacing missing values by specific values

# Calculate the mean of the 'bmi' column, excluding NaN values
mean_bmi = df['bmi'].mean()

# Replace missing 'bmi' values with the calculated mean
df['bmi'].fillna(mean_bmi, inplace=True)

# Check if there are any remaining missing values
print("\nMissing values after imputation:")
print(df.isnull().sum())
```

```python
import pandas as pd
# Recoding 'stroke' variable: 0 for no stroke, 1 for stroke
df['stroke_binary'] = df['stroke'].apply(lambda x: 1 if x == 1 else 0)
print(df['stroke_binary'].value_counts())
```

```python
# Example: Recoding 'age' into categories
# Define age bins and labels
age_bins = [0, 18, 35, 60, 100]
age_labels = ['0-17', '18-34', '35-59', '60+']

# Create the new categorical 'age_group' variable
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Display the value counts of the new 'age_group' variable
print("\nValue counts for recoded 'age_group' variable:")
print(df['age_group'].value_counts())
```
