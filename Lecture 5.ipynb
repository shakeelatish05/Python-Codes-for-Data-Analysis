```python
import pandas as pd
df=pd.read_csv('Strokes.csv')
df.head()
```

```python
# prompt: box plot for average glocose level by gender

import plotly.express as px

fig = px.box(df, x="gender", y="avg_glucose_level")
fig.show()
```

```python
# prompt: Describe avg glucose level by gender  and standard deviation

df.groupby('gender')['avg_glucose_level'].agg(['mean', 'std'])
```

```python
#Applying independent sample t test to compare glocose level between male and female

from scipy import stats

# Assuming 'gender' is the column for gender and 'avg_glucose_level' for glucose level
male_glucose = df[df['gender'] == 'Male']['avg_glucose_level']
female_glucose = df[df['gender'] == 'Female']['avg_glucose_level']


# Perform independent sample t-test
t_stat, p_value = stats.ttest_ind(male_glucose, female_glucose)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpret the results (common significance level is 0.05)
if p_value < 0.05:
  print("Reject the null hypothesis: There is a significant difference in average glucose level between males and females.")
else:
  print("Fail to reject the null hypothesis: There is no significant difference in average glucose level between males and females.")
```

```python
# Welch test Assuming Variances are not equal

# Perform Welch's t-test (independent samples, assumes unequal variances)
welch_t_stat, welch_p_value = stats.ttest_ind(male_glucose, female_glucose, equal_var=False)

print(f"Welch's T-statistic: {welch_t_stat}")
print(f"Welch's P-value: {welch_p_value}")

# Interpret the results for Welch's t-test
if welch_p_value < 0.05:
  print("Reject the null hypothesis (Welch): There is a significant difference in average glucose level between males and females.")
else:
  print("Fail to reject the null hypothesis (Welch): There is no significant difference in average glucose level between males and females.")
```

```python

# Applying independent sample t test for multiple variables (age, bmi, glucose) at the same time

variables_to_test = ['age', 'bmi', 'avg_glucose_level']
gender_groups = df['gender'].unique() # Assuming 'gender' is the grouping variable

for var in variables_to_test:
    print(f"\nPerforming t-test for: {var}")
    group1_data = df[df['gender'] == gender_groups[0]][var].dropna()
    group2_data = df[df['gender'] == gender_groups[1]][var].dropna()

    # Perform independent sample t-test (assuming equal variances)
    t_stat, p_value = stats.ttest_ind(group1_data, group2_data)

    print(f"  Independent samples t-test (equal variances assumed):")
    print(f"    T-statistic: {t_stat}")
    print(f"    P-value: {p_value}")

    if p_value < 0.05:
        print(f"    Reject the null hypothesis: There is a significant difference in {var} between the groups.")
    else:
        print(f"    Fail to reject the null hypothesis: There is no significant difference in {var} between the groups.")
```

```python
#Paired sample t test Average Glocose and Average Glocose 2

# Create two hypothetical paired samples for demonstration
import numpy as np
# For this example, we will simulate a second glucose measurement.
df['avg_glucose_level_2'] = df['avg_glucose_level'] + np.random.normal(3, 5, size=len(df))

# Drop rows with NaN values in either column before performing the test
paired_data = df[['avg_glucose_level', 'avg_glucose_level_2']].dropna()

# Perform the paired sample t-test
t_stat_paired, p_value_paired = stats.ttest_rel(paired_data['avg_glucose_level'], paired_data['avg_glucose_level_2'])

print(f"\nPaired Sample T-test:")
print(f"  T-statistic: {t_stat_paired}")
print(f"  P-value: {p_value_paired}")

# Interpret the results for the paired sample t-test
if p_value_paired < 0.05:
  print("  Reject the null hypothesis: There is a significant difference between the paired measurements of average glucose level.")
else:
  print("  Fail to reject the null hypothesis: There is no significant difference between the paired measurements of average glucose level.")
```
