```python
import pandas as pd
df=pd.read_csv('Strokes.csv')
df.head()
```

```python
# Generate one-way tables for categorical variables
from tabulate import tabulate
categorical_var = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

for var in categorical_var:
    counts = df[var].value_counts()
    frequencies = df[var].value_counts(normalize=True) * 100

    # Create a DataFrame for the table
    table_data = pd.DataFrame({'Frequency': counts, 'Percentage': frequencies})
    print(tabulate(table_data, headers='keys'))
```

```python
##Two way tabulation
print("\nTwo-Way Table with Row Percentages for: gender and ever_married")
two_way_table_pct_row = pd.crosstab(df['gender'], df['ever_married'], normalize='index') * 100
print(tabulate(two_way_table_pct_row, headers='keys', tablefmt='psql'))
```

```python
# prompt: Fisher's exact test

from scipy.stats import fisher_exact
two_way_table = pd.crosstab(df['gender'], df['ever_married'])
# Example: Fisher's Exact Test for 'gender' and 'ever_married'
# We use the two-way table calculated previously
print("\nFisher's Exact Test for: gender and ever_married")
odds_ratio, p_value = fisher_exact(two_way_table)

print(f"Odds Ratio: {odds_ratio}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Result: There is a statistically significant association between gender and marital status (reject the null hypothesis).")
else:
    print("Result: There is no statistically significant association between gender and marital status (fail to reject the null hypothesis).")
```
